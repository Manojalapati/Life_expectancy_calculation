### 5) BMI

### 6) Polio

### 8) Diphtheria

### 12) Income Composition

# Model Building

```python
x = df.iloc[:,df.columns != 'Life_expectancy']
y = df['Life_expectancy']
```

```python
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=16)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

## 1) Linear Regression

```python
from sklearn.linear_model import SGDRegressor
reg_model = SGDRegressor()
reg_model.fit(x_train,y_train)
```

```python
y_pred = reg_model.predict(x_test)
score = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
print("R-Score : ",score)
print("MSE = ",mse)
```

```python
plt.scatter(y_test,y_pred)
```

## 2) XGBOOST

```python
from xgboost import XGBRegressor
xgb_model = XGBRegressor()
xgb_model.fit(x_train,y_train)
```

```python
y_pred_xgb = xgb_model.predict(x_test)
score = r2_score(y_test,y_pred_xgb)
mse = mean_squared_error(y_test,y_pred_xgb)
print("R-Score : ",score)
print("MSE = ",mse)
```

```python
plt.scatter(y_test,y_pred_xgb)
```

## 3) Artificial Neural Network

```python
import tensorflow as tf
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential([
    Dense(units = 10, activation = 'relu'),
    Dense(units = 1, activation = 'linear')
])
```

```python
from keras.losses import MeanSquaredError
model.compile(
    loss=MeanSquaredError(),
    optimizer=tf.keras.optimizers.Adam(0.01),
)

model.fit(
    x_train,y_train,
    epochs=200
)
```

```python
y_pred_ann = model.predict(x_test)
score = r2_score(y_test,y_pred_ann)
mse = mean_squared_error(y_test,y_pred_ann)
print("R-Score : ",score)
print("MSE = ",mse)
```

```python
plt.scatter(y_test,y_pred_ann)
```
